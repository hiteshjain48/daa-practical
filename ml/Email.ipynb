# cleaning

```python
#dropping the column because it is not necessary
df.drop(columns=['Email No.'], inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.drop(['Prediction'],axis = 1)
y = df['Prediction']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#way to control how warning messages are displayed and handled
import warnings
warnings.filterwarnings('ignore')

#importing the required packages
from sklearn.model_selection import train_test_split

#svm model is basically used for classification tasks in machine learning
from sklearn.svm import SVC

#used for evaluating the machine learning models : accuracy score, confusion ,matrix etc
from sklearn import metrics
```

```python
#helps in scalimg and standarizing your data
from sklearn.preprocessing import scale
X = scale(X)

# split into train and test
#randomization number = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
#classifies data points based on majority class among the nearest neighbours
from sklearn.neighbors import KNeighborsClassifier

#considering the lael of nearest 7 neighbours
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print("Prediction",y_pred)
```

```python
y_pred
```

```python
#predicting the accuracy
print("KNN accuracy = ",metrics.accuracy_score(y_test,y_pred))
```

```python
print("Confusion matrix",metrics.confusion_matrix(y_test,y_pred))
```

```python

# svm separates data points in differrent classes in the feature space
model = SVC(C = 1)

# fitting the model
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)
```

```python
metrics.confusion_matrix(y_true=y_test, y_pred=y_pred)
```

```python
print("SVM accuracy = ",metrics.accuracy_score(y_test,y_pred))
```
